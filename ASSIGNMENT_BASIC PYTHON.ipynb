```python
#prime numbers

num=int(input("Enter a number ðŸ˜Š:"))
if num <= 1:
    print("Invalid number")
else:

    is_prime=True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime=False
            break

    if is_prime:
        print(num, "is a Prime Number")
    else:
        print(num,"is not a prime number")
        


        
        
            
   
```

```python
#hospitalBILL
ROOM_CHARGE_PER_DAY = 2000
DOCTOR_FEE = 1500
LAB_TEST_CHARGE_PER_TEST = 300
DISCOUNT_POINT = 10000
DISCOUNT_RATE = 0.10  

# Given Patient details
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

room_charges = days_stayed * ROOM_CHARGE_PER_DAY
lab_test_charges = num_lab_tests * LAB_TEST_CHARGE_PER_TEST

total_before_discount = room_charges + DOCTOR_FEE + lab_test_charges + medicine_charges

discount = 0
if total_before_discount > DISCOUNT_POINT:
    discount = total_before_discount * DISCOUNT_RATE

final_bill = total_before_discount - discount

print("Hospital Bill")
print(f"Room charges ({days_stayed} days @ â‚¹{ROOM_CHARGE_PER_DAY}): â‚¹{room_charges}")
print(f"Doctor consultation fee: â‚¹{DOCTOR_FEE}")
print(f"Lab test charges ({num_lab_tests} tests @ â‚¹{LAB_TEST_CHARGE_PER_TEST}): â‚¹{lab_test_charges}")
print(f"Medicine charges: â‚¹{medicine_charges}")
print(f"Total before discount: â‚¹{total_before_discount}")
print(f"Discount: â‚¹{discount}")
print(f"Final bill: â‚¹{final_bill}")
```
